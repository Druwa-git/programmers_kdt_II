# Problem 2

## Mandatory Part

```python
bitcoin = pd.read_csv('./BitCoin.csv', parse_dates=["Date"])
start_date = datetime.date(2016,6,1)
end_date = datetime.date(2017,7,1)
bitcoin = bitcoin.sort_values(by="Date")
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("BitCoin Contour")
plt.xlim([start_date, end_date])
plt.grid(True)

plt.plot(bitcoin["Date"], bitcoin["Open"].rolling(window=5).mean(), label="Bitcoin", color='#f2a900')
plt.legend()

plt.show()
```

## Bonus Part

```python
etherium = pd.read_csv('./ETH_day.csv', parse_dates=["Date"])
start_date = datetime.date(2016,6,1)
end_date = datetime.date(2017,7,1)
etherium = etherium.sort_values(by="Date")
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("BitCoin and Etherium Contour")
plt.xlim([start_date, end_date])
plt.grid(True)

plt.plot(bitcoin["Date"], bitcoin["Open"].rolling(window=5).mean(), label="Bitcoin", color='#f2a900')
plt.plot(etherium["Date"], etherium["Open"].rolling(window=5).mean(), label="Etherium", color='#3c3c3d')
plt.legend()

plt.show()
```
